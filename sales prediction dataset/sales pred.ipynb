```python
def age_groups_creation(row):
    if row['age']< 30:
        return '20-30'
    elif row['age']<40:
        return '30-40'
    elif row['age']<50:
        return '40-50'
    elif row['age']<60:
        return '50-60'
    else:
        return '60-70'
```

```python
df['Age Ranges']=df.apply(age_groups_creation, axis=1)
```

```python
X=df[['age','annual Salary','net worth']]
y=df[['car purchase amount']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler, 'scaler.pkl')
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
def modelresult(predictions):
    print('Mean Absolute error on model is {} '. format( mean_absolute_error(y_test,predictions)))
    print('Mean Squared error on model is {} '. format( mean_squared_error(y_test,predictions)))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
modelresult(y_pred)
```

```python
df['car purchase amount'].describe()
```

```python
lr.coef_
```

```python
from sklearn.svm import SVR
svrmodel= SVR()
param_gridsvr={'C':[0.01, 0.1, 0.5], 'kernel':['linear','rbf','poly'], 'degree':[2,3,4]}
```

```python
from sklearn.model_selection import GridSearchCV
gridsvr= GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(X_train,y_train.values.ravel())
```

```python
pred_gridsvr=gridsvr.predict(X_test)
```

```python
modelresult(pred_gridsvr)
```

```python
gridsvr.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel= RandomForestRegressor()
param_gridrfr = {'max_depth':[5,10,15], 'n_estimators':[11,12,10,13,14,15,16,17]}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train,y_train.values.ravel())
```

```python
predictionrfr= gridrfr.predict(X_test)
```

```python
modelresult(predictionrfr)
```
